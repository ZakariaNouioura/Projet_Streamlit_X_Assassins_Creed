```python


## Sellenium 
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

# Divers à trier !!
from urllib.parse import urlparse 
from tqdm import tqdm
import time

# Beautifulsoup 
from bs4 import BeautifulSoup as bs
import requests

#pandas
import pandas as pd 
```

```python
url="https://www.metacritic.com/search/game/assassin%27s%20creed/results"

#initialisation de sellenium 
driver = webdriver.Chrome()
driver.get(url)

#initialisation soup

user_agent = {'User-agent': 'Mozilla/5.0'}
response = requests.get(url, headers = user_agent)

#print(response)
soup = bs(response.text, 'html.parser')
last_page=soup.find('li',class_="page last_page").a.text


## ferme la page cookies
try:
    time.sleep(1)
    privacy=driver.find_element(By.ID,"onetrust-accept-btn-handler")
    privacy.click()
except:
    pass

#trouve tous les jeux assassins de la page
games=driver.find_elements(By.PARTIAL_LINK_TEXT, "Assassin's Creed")

for i in range(1,len(games)):
    games[i].click()
    time.sleep(1)
    driver.get(driver.current_url+"/user-reviews")
    print(driver.current_url)
    time.sleep(1)
    driver.back()
    time.sleep(2)
    driver.back()

driver.close
```

```python
list_of_links=[#Assassin's creed :
"https://www.metacritic.com/game/playstation-3/assassins-creed/user-reviews",
"https://www.metacritic.com/game/xbox-360/assassins-creed/user-reviews",
"https://www.metacritic.com/game/xbox-360/assassins-creed-ii/user-reviews",
"https://www.metacritic.com/game/playstation-3/assassins-creed-ii/user-reviews",
#Assassin's creed Brotherhood :
"https://www.metacritic.com/game/playstation-3/assassins-creed-brotherhood/user-reviews",
"https://www.metacritic.com/game/xbox-360/assassins-creed-brotherhood/user-reviews",
#Assassin's creed revelations :
"https://www.metacritic.com/game/xbox-360/assassins-creed-revelations/user-reviews",
"https://www.metacritic.com/game/playstation-3/assassins-creed-revelations/user-reviews",
#Assassin’s creed 3 :
"https://www.metacritic.com/game/playstation-3/assassins-creed-iii/user-reviews",
"https://www.metacritic.com/game/xbox-360/assassins-creed-iii/user-reviews",
#Assassin’s creed Black flag :
"https://www.metacritic.com/game/playstation-4/assassins-creed-iv-black-flag/user-reviews",
"https://www.metacritic.com/game/playstation-3/assassins-creed-iv-black-flag/user-reviews",
"https://www.metacritic.com/game/xbox-360/assassins-creed-iv-black-flag/user-reviews",
"https://www.metacritic.com/game/xbox-one/assassins-creed-iv-black-flag/user-reviews",
#Assassin’s creed Rogue :
"https://www.metacritic.com/game/playstation-3/assassins-creed-rogue/user-reviews",
"https://www.metacritic.com/game/xbox-360/assassins-creed-rogue/user-reviews",
#Assassin's Creed Unity:
"https://www.metacritic.com/game/playstation-4/assassins-creed-unity/user-reviews",
"https://www.metacritic.com/game/xbox-one/assassins-creed-unity/user-reviews",
#Assassin's Creed Syndicate :
"https://www.metacritic.com/game/playstation-4/assassins-creed-syndicate/user-reviews",
"https://www.metacritic.com/game/xbox-one/assassins-creed-syndicate/user-reviews",
#Assassin's Creed Origins :
"https://www.metacritic.com/game/playstation-4/assassins-creed-origins/user-reviews",
"https://www.metacritic.com/game/xbox-one/assassins-creed-origins/user-reviews",
#Assassin's Creed Odyssey :
"https://www.metacritic.com/game/playstation-4/assassins-creed-odyssey/user-reviews",
"https://www.metacritic.com/game/xbox-one/assassins-creed-odyssey/user-reviews",
#Assassin's Creed Valhalla :
"https://www.metacritic.com/game/playstation-5/assassins-creed-valhalla/user-reviews",
"https://www.metacritic.com/game/playstation-4/assassins-creed-valhalla/user-reviews",
"https://www.metacritic.com/game/xbox-one/assassins-creed-valhalla/user-reviews",
"https://www.metacritic.com/game/xbox-series-x/assassins-creed-valhalla/user-reviews"]

L_comment=[]
L_evaluation=[]
L_date=[]
L_name=[]

for link in list_of_links :

    user_agent = {'User-agent': 'Mozilla/5.0'}
    response = requests.get(link, headers = user_agent)

    print(response,link)
    #print(response)
    soup = bs(response.text, 'html.parser')

    ## Recupère la derniere page

    ##
    try:
        last_page=pages.find("li",class_="page last_page").a.text
    except AttributeError:
        last_page=None
    
    # Si il y a plusieurs pages d'avis :
    if soup.find("div",class_="pages"):
        pages=soup.find("div",class_="pages")
        print("Il y a plusieurs pages")
        last_page=int(pages.find("li",class_="page last_page").a.text)
        for page in range(last_page):
            link_page=link+'?page='+str(page)
            print(link_page)
            #response_dynamic = requests.get(link_page, headers = user_agent)
            #soup_dynamic = bs(response.text, 'html.parser')
            reviews=soup.find_all("div",class_="review_content")
            for review in reviews :
                try:
                    comment=review.find("div",class_="review_body").span.text 

                except AttributeError:
                    comment=None
                L_comment.append(comment)
                try: 
                    evaluation=review.find("div",class_="metascore_w user medium game positive indiv perfect").text
                except AttributeError:
                    evaluation=None
                L_evaluation.append(evaluation)
                try:
                    date=review.find("div",class_="date").text
                except AttributeError: 
                    date=None
                L_date.append(date)   
                try: 
                    name=soup.find("div",class_="product_title").h1.text.strip()
                except AttributeError: 
                    name=None
                L_name.append(name)

    #Si il y a qu'une seule page d'avis :   
    else:
        print("Il y a qu'une seul page")
        reviews=soup.find_all("div",class_="review_content")

        for review in reviews :
            try:
                comment=review.find("div",class_="review_body").span.text  
            except AttributeError:
                comment=None
            L_comment.append(comment)
            try: 
                evaluation=review.find("div",class_="metascore_w user medium game positive indiv perfect").text
            except AttributeError:
                evaluation=None
            L_evaluation.append(evaluation)
            try:
                date=review.find("div",class_="date").text
            except AttributeError: 
                date=None
            L_date.append(date)
            try: 
                name=soup.find("div",class_="product_title").h1.text.strip()
            except AttributeError: 
                name=None
            L_name.append(name)
```

```python
print(len(L_comment),len(L_date),len(L_evaluation))
df=pd.DataFrame(list(zip(L_evaluation,L_date,L_comment,L_name)),columns=["note",'date_commentaire',"Commentaire",'nom_du_jeu'])
df.to_excel("Metacritic_Assassins_Creed_comments.xlsx")
df.shape
df.tail(60)
```
